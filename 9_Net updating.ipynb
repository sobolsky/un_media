```python
#import numpy as np
#import pandas as pd
import networkx as nx
#from matplotlib.pyplot import figure
```

```python
G = nx.read_gexf('UN_UKRAINE--POLITICAL_CONDITIONS_attr_2023.gexf')
```

```python
G.nodes(data=True)
```

```python
vote_arr = ['-1-1-1-1-1-1',
 '00-10-10',
 '00-1000',
 '000000',
 '100101',
 '110001',
 '110101',
 '110111',
 '111101',
 '111111']

'-1-1-1-1-1-1' in vote_arr
```

```python
# add new field in nodes
for node in G.nodes:
    if G.nodes[node]['votes_profile'] in vote_arr:
        G.nodes[node]['sol_group'] = G.nodes[node]['votes_profile']
    else:
        G.nodes[node]['sol_group'] = 'none'
```

```python
G.nodes(data=True)
```

```python
# Обход всех ребер и преобразование значения атрибута 'weight' в тип 'int'
#for u, v, data in G.edges(data=True):
#    data['weight'] = str(data['weight'])
```

```python
# добавление нового поля в ребра
for u, v, data in G.edges(data=True):
    data['weight_int'] = int(data['weight'])
```
