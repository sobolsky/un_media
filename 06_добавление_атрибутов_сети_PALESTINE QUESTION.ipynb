# 06_добавление_атрибутов_сети_профиль_голосования

```python
import numpy as np
import pandas as pd
import networkx as nx
from matplotlib.pyplot import figure
```

```python
topic = 'PALESTINE QUESTION'
```

## 1. Добавление координат в сеть

```python
# страны исправлены
vote_table = pd.read_csv(topic + '_vote_table.csv')
#vote_table = pd.DataFrame(vote_table.iloc[:, [0, 5, 4, 3, 2, 1]])
```

### Votes profile creating

```python
vote_table
```

```python
for column_caption in list(vote_table.columns)[1:]:
       vote_table[column_caption] = [str(x)[:-2] for x in vote_table[column_caption]]
        
vote_table
#vote_table[vote_table['Country'] == 'TÜRKIYE']
```

```python
vote_table['Votes profile'] = vote_table.iloc[:,1:].sum(axis = 1)
```

```python
vote_table
#vote_table[vote_table['Country'] == 'TÜRKIYE']
```

### The Net updating

```python
nodes_array = []

for index in range (0, len(vote_table)):
    nodes_array.append(
        (
            vote_table['Country'][index], 
            {'votes_profile': vote_table['Votes profile'][index]}
        ) 
    )
```

```python
G = nx.Graph()
G.add_nodes_from(nodes_array)
len(G)
```

```python
Contry_att_net = nx.read_gexf('UN_UKRAINE--POLITICAL_CONDITIONS_attr_4.gexf')
Contry_att_net.nodes(data=True)
```

```python
G2 = nx.read_gexf('UN_PALESTINE_QUESTION.gexf')
G2.add_nodes_from(Contry_att_net.nodes(data=True))
```

```python
G2.nodes(data=True) 
```

```python
R = nx.compose(G, G2)
```

```python
R.nodes(data=True)
```

```python
R.edges(data=True)
```

```python
nx.write_gexf(R, 'UN_PALESTINE_QUESTION_attr_2.gexf')
```

### country_attr_table_3 updating

```python
country_attr_table = pd.read_csv('country_attr_table_3.csv')
country_attr_table.drop(columns = {'Unnamed: 0', 'Unnamed: 0.1'}, inplace = True)
#country_attr_table[country_attr_table['Country'] == 'TÜRKIYE']
country_attr_table
```
